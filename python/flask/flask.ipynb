************************URL BUILDING************************

the **url_for()** function is used to build a URL to the specific function dynmaically 

the first argument is the name of the specified fucntion and then we pass any no of keyword argumnets corresponding to the variable part to the url



**redirect()**
used to send automatically send the user to a differenct route(url)

**url_for**
used to automaticaly generate the routr(url) using the function name,instead of hardcoding the path


```python
# **************code for url building****************
from flask import Flask,url_for,redirect
app=Flask(__name__)

@app.route("/admin")
def admin():
    return "this is admin"

@app.route("/student")
def student():
    return "this is student"

@app.route("/staff")
def staff():
    return "this is staff"

@app.route('/user/<name>')
def user(name):
    if name=="admin":   #here we should pass the the url endpoint
        return redirect(url_for("admin"))  # you should padd the fucnion not the url
    if name=="student":
        return redirect(url_for('student'))
    if name=='staff':
        return redirect(url_for('staff'))

if __name__=='__main__':     # Only run the code below if this script is executed directly (not imported)
    app.run()
```

*******************App Routing*****************
App Routing in Flask is the technique of mapping URLs to fucntions.

App routing is used to map the specific URL with the associated fucntion that is intended to perform some task. it is usd to access some particular page

in other words, we can say that if we visit the particular URL mapped to some particular fucntion, the output of that funciton is renderes on the brower's screen.

**@app.route(`/hello`)**


**The add_url_rule()** function
There is one more approch to perform routing for the flask web applications that can be done using the add_url() fucntion of the Flask class. 
        <!-- syntax: add_url_rule(<url rule>,<endpoint>,<view fucntion>) -->

```python
# **How to pass data through url to the fucntion**
from flask import Flask

app = Flask(__name__)

@app.route('/hello/<sub>')
def hello(sub):
    return "Hello!"+sub


if __name__ == "__main__":
    app.run(debug=True)




# we can pass integrers
from flask import Flask

app = Flask(__name__)

@app.route('/hello/<int:sub>')
def hello(sub):
    return "Hello!%d"%sub


if __name__ == "__main__":
    app.run(debug=True)
```

```python
# ****add_url_rule(<url rule>,<endpoint>,<view fucntion>)*******

from flask import Flask

app=Flask(__name__)
# @app.route("/") we can replace this code  with add_url_rule(), we have done below....
def mittu():
    return "i am navaneeth"
app.add_url_rule("/mittu","mittu",mittu)


if __name__=="__main__":
    app.run()
```

*******************Flask Templates************************

Flask used **Jinja2** as its templating engine.

Templates  let you separate HTML design from python logic.

Instead of returning raw HTML as strings from your routes, you can create separate .html files and render them using Flask.

**Delimiters**
Jinja2 tempalte engine provides some delimiters which can be used in the HTML to make it capable of dynamic dat representation.

ex:
        {%....%}        for statements
        {{...}}         for expressions to print to the tempalte output
        {#....#}        for the comments that are not included inthe template putput
        #...##          for line statements



```python
# Delimiters code  python

from flask import *


app=Flask(__name__)


@app.route("/user/<name>")

def msg(name):  # we can give multiple values as parameters but the order pf paraters should be same position which were given in dynamic routes
    return render_template("base.html",name=name)  


if __name__=="__main__":
    app.run(debug=True)
```
