```python
# basic structure of psycopg code


import psycopg2

conn=psycopg2.connect(
    dbname="postgres",
    user="postgres",
    password="Mittu@1821",
    host="localhost",
    port="5432"
)

cur=conn.cursor()

# cursor() is an intermediary object that allow  us to execute sql quries,fetch rsuls and manage transactions.

cur.execute("""select * from postgres_test.users""")

# execute() used to run sql quries, it is an cursor object,which is created by calling cursor() 


rows=cur.fetchall()
for row in rows:
    print(row)


cur.close()
conn.close()
```
