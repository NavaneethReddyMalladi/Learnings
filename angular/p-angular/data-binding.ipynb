******one way data binding*******

1. **Interpolation in Angular**

- Interpolation in Angular is a way to bind data from your component class to your HTML template using double curly braces {{ }}.

<!-- <h1>Hello {{title}}</h1> -->

# interpolaiton can be used with

1. variables  {{tite}}

2. method calls {{getdata()}}

3. arithmetic/expressions  {{count+1}}


2. **property binding**
Property binding in Angular is a way to set values from your component to HTML element properties in the template using square brackets [ ].

- [elementProperty]="componentProperty"

ex-         <input type='text' placeholder="enter" [value]="date" class="form-control"/>

ex-2-        <input [type]="inputtype"/>


3. **event binding**
In Angular, event binding allows you to listen to DOM events and trigger methods in your component. This is similar to handling events in JavaScript but follows Angular's binding syntax.

# events should be written in open braces

<button class="btn btn-success" (click)="showWelcomeAlert()">show alert</button>
<button class="btn btn-success" (click)="showMsg('ok')">show alert alert</button>
